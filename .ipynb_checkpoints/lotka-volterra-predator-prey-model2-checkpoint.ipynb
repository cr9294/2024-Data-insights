在生态系统中，捕食者和猎物之间的相互作用可以用一个简单的模型来捕捉，这就是著名的Lotka-Volterra竞争模型。

设$U(t)$和$V(t)$分别为某一时刻$t$的猎物和捕食者的种群大小。种群随时间的变化由以下一组微分方程描述：
$$
\begin{aligned}
\frac{dU}{dt} &= \alpha U - \gamma UV \\
\frac{dV}{dt} &= - \beta V + e\gamma UV 
\end{aligned}
$$

方程右侧的第一项，即$\alpha$和$\beta$项，描述了如果该物种被单独留下时种群大小的变化趋势。假设猎物始终有充足的食物资源，那么在没有捕食者的情况下，它们的种群大小将增长。另一方面，如果没有猎物可捕食，捕食者的种群大小将减少。这些考虑导致了$\alpha$和$\beta$项前的正负号。

交叉项，即$UV$项，代表了两个物种之间的相互作用。$\gamma$可以被视为每单位时间被捕食者捕获的猎物的比例，因此在$dt$内捕获的猎物总数是$\gamma dt \cdot UV$。对于捕食者来说，捕获更多的猎物意味着它们可以生更多的幼崽。由于可能存在某种转换因子，我们称之为$e$，因此在$dt$内捕食者的新生幼崽数量为$e\gamma dt \cdot UV$。因此，方程右侧的第二项描述了这种猎物-捕食者相互作用。

通过规范化$\tau = \sqrt{\alpha \beta} t$，$u = \frac{e\gamma}{\beta}U$，和$v = \frac{\gamma}{\alpha}V$，Lotka-Volterra方程可以简化为以下形式：
$$
\begin{aligned}
\frac{du}{dt} &= r(u - uv) \\
\frac{dv}{dt} &= \frac{1}{r}(v + uv) 
\end{aligned}
$$

因此，这个模型可以通过一个常数$r$来参数化，而不是四个参数$\alpha$，$\beta$，$\gamma$和$e$，其中
$$
r = \sqrt{\frac{\alpha}{\beta}}
$$

Lotka-Volterra方程是非线性的，没有简单的解析解，尽管它们预测种群大小会呈现定期的周期循环。在接下来的模拟中
，我们将研究Lotka-Volterra方程的数值解。

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
#import matplotlib.pyplot as plt
#plt.rcParams['font.family'] = 'SimHei'  # 替换为你喜欢的字体
```

```python
# define the lotka-volterra prey-predator model
# u = relative number of preys
# v = relative number of predators
# r = dimensionless parameter
def lotka_volterra(uv, t, r):
    u, v = uv
    dudt = r * (u - u*v)
    dvdt = 1/r * (u*v - v)
    return [dudt, dvdt]
```

```python
# 设定要绘制的相位图的数量
step = 0.2
n_step = 7

# 绘制给定r值的相位图
def phase_diagram(r):
    # 设置初始prey数值范围
    u0 = np.arange(1.1, 1.1 + n_step*step, step)
    
    for i in range(n_step):
        # 解Lotka-Volterra方程并绘制相位图
        sol = odeint(lotka_volterra, [u0[i], 1.1], t, args=(r,))
        plt.plot(sol[:, 0], sol[:, 1], 'black')
    
    # 设置图表标签和标题
    #preg
    plt.xlabel('prey')
    plt.ylabel('predator')
    plt.title('r = ' + str(r), loc='center')
    plt.ylim([0, 4.5])
    plt.grid()
    plt.show()
```

```python
# 绘制给定r值下猎物和捕食者数量的变化
def time_var(r, initial_uv):
    # 解Lotka-Volterra方程并绘制数量随时间的变化图
    sol = odeint(lotka_volterra, initial_uv, t, args=(r,))
    plt.plot(t, sol[:, 0], 'green', label='prey')
    plt.plot(t, sol[:, 1], 'red', label='predator')
    plt.xlabel('time')
    plt.ylabel('number')
    plt.ylim([0.4, 1.6])
    plt.legend(loc='upper right')
    plt.title('r = ' + str(r), loc='center')
    plt.grid()
    plt.show()
```

```python
# set up duration for the simulation
#设置模拟持续时间
t = np.linspace(0, 10, 200)
```

```python
phase_diagram(0.5)
```

```python
phase_diagram(1.0)
```

```python
phase_diagram(2.0)
```

```python
# 设置模拟的时间范围
t = np.linspace(0, 16, 300)
# 调用time_var函数绘制r值为0.5时猎物和捕食者数量的变化图
time_var(0.5, [1.2, 1.0])
```

```python
time_var(1.0, [1.2, 1.0])
```

```python
time_var(2.0, [1.2, 1.0])
```

```python
# 比较给定初始u和v值的不同r值下的相位图
r_candidates = [0.5, 0.7, 1.0, 1.4, 2.0]
t = np.linspace(0, 10, 200)

# 定义compare_r函数
def compare_r(initial_uv):
    # 遍历不同r值
    for r in r_candidates:
        # 解Lotka-Volterra方程并绘制相位图
        sol = odeint(lotka_volterra, initial_uv, t, args=(r,))
        plt.plot(sol[:, 0], sol[:, 1], label='r = '+str(r))
    
    plt.xlabel('prey')
    plt.ylabel('predator')
    plt.legend(loc='upper right')
    plt.grid()
    plt.show()
```

```python
# 使用compare_r函数比较给定初始u和v值 [1.2, 1.0] 的不同r值下的相位图
compare_r([1.2, 1.0])
```

```python
compare_r([1.0, 1.2])
```
