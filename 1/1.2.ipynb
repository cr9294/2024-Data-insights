```python
import matplotlib.pyplot as plt
import numpy as np

def simulate_gender_ratio_effect(initial_gender_ratio, food_supply_factor):
    generations = 50
    gender_ratio = np.linspace(initial_gender_ratio, 1 - initial_gender_ratio, generations)
    food_supply = food_supply_factor / gender_ratio

    # 绘制折线并在折线下方添加阴影
    plt.plot(gender_ratio, food_supply, label=f'Initial Gender Ratio: {initial_gender_ratio:.2f}', linewidth=2)
    plt.fill_between(gender_ratio, 0, food_supply, alpha=0.3)

# 模拟性别比例从全雌性到全雄性的变化
initial_ratios = [0.1, 0.3, 0.5, 0.7, 0.9]
food_supply_factor = 100  # 初始食物资源数量

# 设置图的风格
plt.style.available
plt.style.use('seaborn-darkgrid')#有什么风格可以选，可以查看plt.style.available

plt.figure(figsize=(10, 6))
for initial_ratio in initial_ratios:
    simulate_gender_ratio_effect(initial_ratio, food_supply_factor)

plt.title('Effect of Gender Ratio on Food Supply', fontsize=16)
plt.xlabel('Gender Ratio', fontsize=14)
plt.ylabel('Food Supply', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
