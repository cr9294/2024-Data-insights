```python
import matplotlib.pyplot as plt
import numpy as np
```


1. 初始雄性比例：
   \[
   \text{initial\_ratio} = 
   \begin{cases} 
   0.22 & \text{if } \text{food\_availability} = \text{'low'} \\
   0.56 & \text{if } \text{food\_availability} = \text{'high'}
   \end{cases}
   \]

2. 雄性比例变化过程：
   \[
   \text{gender\_ratio}[gen] = 
   \begin{cases} 
   \text{gender\_ratio}[gen-1] + 0.01 & \text{if } \text{food\_availability} = \text{'low'} \\
   \text{gender\_ratio}[gen-1] - 0.01 & \text{if } \text{food\_availability} = \text{'high'}
   \end{cases}
   \]

这里，\(gen\) 表示世代数，\(\text{gender\_ratio}[gen]\) 表示第 \(gen\) 代的雄性比例。函数返回一个包含 50 个元素的雄性比例数组。

```python
# 用法
def simulate_gender_ratio(food_availability):
    # 模拟七鳃鳗雄性比例的变化
    if food_availability == 'low':
        initial_ratio = 0.22  # 假设初始雄性比例为 22%
    elif food_availability == 'high':
        initial_ratio = 0.56  # 假设初始雄性比例为 56%
    else:
        raise ValueError("Invalid food availability")
    
    generations = 50 #与七鳃鳗的寿命相对应，和这一行相关联generations = np.arange(50)
    gender_ratio = np.zeros(generations)
    gender_ratio[0] = initial_ratio
    for gen in range(1, generations):
        # 模拟雄性比例的变化，根据食物可得性调整
        if food_availability == 'low':
            gender_ratio[gen] = gender_ratio[gen - 1] + 0.01  # 在食物较少的环境中，雄性比例增加
        elif food_availability == 'high':
            gender_ratio[gen] = gender_ratio[gen - 1] - 0.01  # 在食物较多的环境中，雄性比例减少
    return gender_ratio
```

```python
# 模拟食物可得性较低的情况
low_food_gender_ratio = simulate_gender_ratio('low')
# 模拟食物可得性较高的情况
high_food_gender_ratio = simulate_gender_ratio('high')
print(low_food_gender_ratio)
print(high_food_gender_ratio)
generations = np.arange(50)
```

```python
# 绘制结果
import numpy as np

# 假设你已经有了数据：generations, low_food_gender_ratio, high_food_gender_ratio
# 创建图形，并设置大小
plt.figure(figsize=(10, 6))
# 绘制图表
plt.plot(generations, low_food_gender_ratio, label='Low Food Availability', linestyle='-', marker='o', linewidth=2)
plt.plot(generations, high_food_gender_ratio, label='High Food Availability', linestyle='--', marker='s', color='green', linewidth=2)

# 添加标题和标签
plt.title('Simulation of Gender Ratio in Seven-Gilled Shark', fontsize=16)
plt.xlabel('Generations', fontsize=12)
plt.ylabel('Gender Ratio', fontsize=12)

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.7)

# 设置 x 轴标签顶格
plt.margins(x=0.021)
# 添加图例并设置位置
plt.legend(loc='lower left')
# 显示图形
plt.show()
```
